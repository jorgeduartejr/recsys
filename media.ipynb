```python
import os
import pandas as pd 
import numpy as np
```

```python
# importando os dados
data = pd.read_csv('u.data',sep='\t', names=['UserID', 'MovieID', 'Rating', 'Timestamp'])


df = data.drop('Timestamp', axis=1) 

print(df.head())
```

```python

media_ratings = data['Rating'].mean()

print('A média das avaliações é: ', media_ratings)

# top 10 filmes melhores avaliados

top_10 = df.groupby('MovieID')['Rating'].mean().sort_values(ascending=False).head(10)

print('Top 10 filmes melhores avaliados: \n', top_10)

```

```python
# filtrando com maior média e maior número de avaliações

# colocar em ordem de maior nota e maior número de avaliações


df2 = pd.read_csv('u.item', sep='|', encoding='latin-1', names=['MovieID', 'Title'], usecols=range(2))

df = df.groupby('MovieID').mean('Rating').sort_values(['Rating'], ascending=False)

df = df.join(df2,on=['MovieID'])


display(df)


# agrupando os que tem maior rating e maior número de avaliações, ordem decrescente

df = data.groupby('MovieID').agg({'Rating': [np.size, np.mean]})

display(df)


```
