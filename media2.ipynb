```python
import pandas as pd

# Carregar dados do arquivo CSV
df = pd.read_csv('u.item', sep='|', encoding='latin-1', names=['MovieID', 'Title'], usecols=range(2))

# Ler os dados de avaliações
ratings_df = pd.read_csv('u.data', sep='\t', names=['UserID', 'MovieID', 'Rating'], usecols=range(3))

# Calcular a média e o número de avaliações por filme
df_stats = ratings_df.groupby('MovieID').agg({'Rating': [pd.Series.count, pd.Series.mean]}).reset_index()
df_stats.columns = ['MovieID', 'RatingCount', 'AvgRating']

display(df_stats.head())

# Juntar os dados dos filmes com as estatísticas
df_stats = pd.merge(df_stats, df, on='MovieID')

# Filtrar filmes com mais de 100 avaliações

df_stats = df_stats[df_stats['RatingCount'] > 100]

# Ordenar os filmes por avaliação média

df_stats = df_stats.sort_values(by='AvgRating', ascending=False)

display(df_stats.head(10))

# ordenar os que tem maior avaliação média

df_stats = df_stats.sort_values(by='RatingCount', ascending=False)
print("Filmes que possuem a maior avaliação média")
display(df_stats.head(10))
```
